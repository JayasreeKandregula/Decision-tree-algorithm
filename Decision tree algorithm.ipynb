```python
from sklearn.datasets import load_iris
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X=load_iris().data
y=load_iris().target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.4, random_state=0)
```

```python
decisionClassifier = DecisionTreeClassifier()
```

### Functions used in decision tree
#### criterion: 
    gini, entropy, log_loss -- default: gini
#### max_depth: 
    maximum depth of the tree -- default: None (any integer) 
#### min_samples_split: 
    The minimum number of samples required to split an internal node -- default:2
#### min_samples_leaf:
    minimum samples at leaf nodes. default-1
#### max_leaf_nodes:
    maximum leaf nodes, default: None


```python
decisionClassifier.fit(X_train,y_train)
```

```python
#plot the decision tree
fig, ax = plt.subplots(figsize=(7,7))
tree.plot_tree(decisionClassifier,ax=ax)
plt.show()
```

```python
y_pred=decisionClassifier.predict(X_test)
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```

```python
metrics.confusion_matrix(y_test, y_pred)
```

```python
metrics.recall_score(y_test, y_pred,average='weighted')
```
